```python
import numpy as np 
import cv2
from scipy.signal import convolve2d
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('cat_.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
def show_differences(kernel):
    convolved = convolve2d(image, kernel)
    fig = plt.figure(figsize=(15,15))
    plt.subplot(121)
    plt.title('Original image')
    plt.axis('off')
    plt.imshow(image, cmap='gray')

    plt.subplot(122)
    plt.title('Convolved image')
    plt.axis('off')
    plt.imshow(convolved, cmap='gray')

    return convolved
```

```python
kernel = np.array([[1,1,1], [1,1,1], [1,1,1]])/9 # box blur
output = show_differences(kernel)
```

```python
kernel = np.array([[1,2,1], [2,4,2], [1,2,1]])/16 #Gaussian blur 3x3
output = show_differences(kernel)
```

```python
kernel = np.array([[-1,-1,-1], [-1,8,-1], [-1,-1,-1]]) # for vertical edges detection
output = show_differences(kernel)
```

```python
kernel = np.array([[-1,1,-1], [-1,1,-1], [-1,1,-1]])
output = show_differences(kernel)
```

```python
kernel = np.array([[0,-1,0], [-1,5,-1], [0,-1,0]]) # sharpen
output = show_differences(kernel)
```
